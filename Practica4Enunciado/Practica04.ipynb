```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from utils import load_data, load_weights, one_hot_encoding, accuracy, displayData
from MLP import MLP
from public_test import compute_cost_test, predict_test


x,y = load_data('data/ex3data1.mat')
theta1, theta2 = load_weights('data/ex3weights.mat')

nn = MLP(theta1.T, theta2.T)
a1,a2,a3,z2,z3 = nn.feedforward(x)
p = np.argmax(a3, axis=1)

predict_test(p, y, accuracy)

compute_cost_test(nn, a3, one_hot_encoding(y))

y = np.where(y != 0, 1, 0)
p = np.where(p != 0, 1, 0)
cm = confusion_matrix(y, p)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()

tp = cm[0][0]
fp = cm[0][1]
fn = cm[1][0]
tn = cm[1][1]

precission = tp / (tp + fp)
print(precission)
recall = tp / (tp + fn)
print(recall)
f1score = 2 * ((precission * recall) / (precission + recall))
print(f1score)
```
